# 爬虫：通过浏览器去获取网页中的一些内容
学习前端知识——了解内容存在什么地方，它的结果是怎样的——分析网页数据，拿到想要的东西

```python
# 爬虫怎么写的综合流程：爬取猫眼上榜单TOP100的电影
'''
分析流程：
1.分析界面/网页的数据来源（数据在界面中哪些位置，需要有翻页功能，翻页功能需要改变的地方在哪），确定请求链接
https://maoyan.com/board/4             -》十个数据
https://maoyan.com/board/4?offset=10   -》十个数据   差异在offset偏移量上面

    (1)先获取一页，在获取多页
2.判断这个链接返回的数据里有没有我想要的数据——检查-network-看请求的url-确定了数据来源的链接
3.写代码请求
4.解析数据 -》拿到我想要的数据
5.将获取的数据进行保存
'''

'''
初始结构：
import requests
class MYSpider(object): 
    pass                           # 对爬虫类对象进行爬虫的操作
if __name__ == "__main__": 
    myspider = MYSpider()
先忽略细节写法看代码流程框架，先写框架，再细节填充
  思考路径：
  base_url = 'https://maoyan.com/board/4?offset={}'      init方法传参要2个
   response = requests.get(url=url,headers= self.headers)
    url=self.base_url.format(start_num)
'''


import requests
# import re

# 定义一个爬虫类
class MYSpider(object):
    def __init__(self,base_url,url_headers):         # class之后，__init__方法，写header之类的。
        # 最基础的URL
        self.base_url = base_url
        self.headers = url_headers
        
    
    # 获取第一页的内容，抓取
    def get_onePage(self,start_num):
        
        # 拼接url
        url=self.base_url.format(start_num)
        # 发送请求
        response = requests.get(url=url,headers= self.headers)
        # 判断是否请求成功，避免程序空跑
        if response.status_code == 200:
            return response.content.decode('utf-8')
        else:
            return None
        
    # 解析数据
    def parse_onePage(self,html):
        '''使用正则表达式解析数据——拿到想要的数据，需要导入re的模块（正则表达式（通过符号匹配出自己想要的内容）的模块）
        知道信息在哪找：检查-elements更方便'''
          # 只取排名
        # pattern = re.compile('<dd>.*?board-index.*?>(\d+)</i>.*?</dd>',re.S)     # compile写正则表达式
        
        # 获取排名和名字
#         pattern = re.compile('<dd>.*?board-index.*?>(\d+)</i>.*?movie-item-info.*?>.*?<a'
#                              +'.*?title="(.*?)".*?>.*?</dd>',re.S)
        # 添加主演
#         pattern = re.compile('<dd>.*?board-index.*?>(\d+)</i>.*?movie-item-info.*?>.*?<a'
#                              +'.*?title="(.*?)".*?>.*?<p class="star">(.*?)</p>.*?</dd>',re.S)
        
        # 上映时间
#         pattern = re.compile('<dd>.*?board-index.*?>(\d+)</i>.*?movie-item-info.*?>.*?<a'
#                              +'.*?title="(.*?)".*?>.*?<p class="star">(.*?)</p>'
#                              +'.*?<p class="releasetime">(.*?)</p>.*?</dd>',re.S)
        # 评分
        pattern = re.compile('<dd>.*?board-index.*?>(\d+)</i>.*?movie-item-info.*?>.*?<a'
                             +'.*?title="(.*?)".*?>.*?<p class="star">(.*?)</p>'
                             +'.*?<p class="releasetime">(.*?)</p>'+
                             '.*?<i class="integer">(.*?)</i>.*?<i class="fraction">(.*?)</i>.*?</dd>',re.S)
        return re.findall(pattern,html)
        
    # 保存数据
    def save_date(self,data):
        for value in data:
            list1 = []
            for valuedata in value:
                # 取出空格
                result  =  valuedata.strip()
                list1.append(result)
            
#             将列表内容链接成字符串
            movestr = " ".join(list1)+'\n'
            print(movestr)
            with open('movestr.txt','a',encoding='utf-8') as f:
                f.write(movestr)

if __name__ == "__main__":             # 如果判断成立，运行下面的那些东西——所以下面的是要作为操作的起步的，然后返回去写语句
    
    
    base_url = 'https://maoyan.com/board/4?offset={}'
    headers = {
        "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.131 Safari/537.36"
    }
    myspider = MYSpider(base_url,headers)
    # 调用获取数据的方法，变量接收数据
    html = myspider.get_onePage(0)
#     print(html)
    
    # 数据传去解析
    resultdata =  myspider.parse_onePage(html)
    myspider.save_date(resultdata)
    print(resultdata)
    
```

## 举例一：拿到网页内容

```python
import requests

# 创建请求头——headers一般都是一个字典：有很多东西，必要加user-agent
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'}

# 发送网络请求
response = requests.get(url='http://www.baidu.com',headers = headers)  # 此处的url可变,get请求方式，两个参数：url，headersv

# 获取请求的内容
# print(response.content.decode('utf-8'))        # 获取变量的内容.content，返回来的都是bytes类型--因为服务器返回来的就是二进制数据bytes，这种方法不容易丢失数据，最常用；如果不编码的话，是b'xxx'的内容，bytes类型
# print(response.text)                   # 另一种获取response里文本的方法，对bytes文件内部做了转换，可能有数据丢失，不建议

# 获取响应头
print(response.headers)

# 状态码
print(response.status_code)
```

## 举例二：获取图片

```python
# 首先要知道图片的链接是什么：找到图片，检查，看源码是否有链接；如果没有，network，找到gif，preview把response的东西进行预览，就是header里面的url

import requests

# 创建请求头
headers = {"User-Agent":'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'} 

#发送网络请求www.baidu.com/img/bd_logo1.png?where=super
response = requests.get(url='https://www.baidu.com/img/bd_logo1.png?where=super',headers = headers)
# print(response.content)

# 将获取的二进制内容进行保存
with open('./baidu.png','wb') as f:
    f.write(response.content)

# f = open('./baidu.png','wb')
# f.write(response.content)
# f.close()
```
